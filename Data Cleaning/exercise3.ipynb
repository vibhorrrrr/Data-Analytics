```python
import pandas as pd

data = {
    'Name':['Ravi', 'Sneha', 'Amit', 'Sneha', None],
    'Age' : ['25', '30', None, '30', '27'],
    'Salary': [60000, 65000, None, 65000, 62000],
    'Department':['IT', 'HR', 'Finance', 'HR', None]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

## Task 1: Count Missing Values

```python
print(df.isnull().sum())
```

## Fill Missing Age with Median

```python
print(df.dtypes)
```

convert Age to Numeric value

```python
df['Age'] = pd.to_numeric(df['Age'], errors = 'coerce')
```

filling the missing age with the median

```python
df['Age'].fillna(df['Age'].median(), inplace = True)
```

```python
print(df)
```

## Task 3: Deleting rows with missing Name (Drop)

```python
df = df.dropna(subset=['Name'])
```

## Task 3: Removing duplicate rows

```python
df = df.drop_duplicates()
```

## Final Task: Filter Employees from HR:

```python
print(df[df['Department'] == 'HR'])
# print df(entry) where 'Department' == HR
```
