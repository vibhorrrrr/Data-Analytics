# **1. What is Data Visualization?**
Data Visualization is a graphical representaion of information/data. It helps **communicate insights, detect trends and simplify complex data**

## **Why is it important?**
- Makes Pattern easier to spot.
- Used in dashboards, reports, and presentations.
- Helps non-technical stakeholders understand data.

# **2. Introduction to Matplotlib**

```python
import matplotlib.pyplot as plt

students = ['Ravi', 'Sneha', 'Arjun', 'Diya', 'Rishi']
scores = [88, 79, 62, 91, 55]
```

## **1. Line Plot (For Trends)**

```python
plt.plot(students, scores) # plot(x, y)
plt.title("Student Scores")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

## **2. Bar Chart**
Great for comparing categories

```python
plt.bar(students, scores, color= 'Skyblue')
plt.title("Student Scores")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

# **2. Introduction to Seaborn**

### Load and preview: 

```python
import seaborn as sns
import pandas as pd

df=pd.read_csv("students.csv")
```

### **1. Histogram**
Shows distribution of scores. Helps detect skewness, central tendency, and outliers.

```python
sns.histplot(df['Score'], bins = 5, kde = True)
# Bins = x axis is divided into bins which help summarize the dist. of data by grouping continuous values into discrete intervals
# KDE = kernal density estimation, reveals whether the data is normally distributed skewed, multimodal or has other patterns
plt.title("Score Distribution")
plt.show()
```

### **2. Countplot(Categorical bar chart)**

```python
sns.countplot(x = 'Course', data = df)
plt.title("Number of Students per Course")
plt.xticks(rotation = 15)
plt.show
```

### **2. Boxplot(Outlier detection + Comparison)**

```python
sns.boxplot(x= 'Course', y='Score', data = df)
plt.title("Score distribution by Course")
plt.show()
```
