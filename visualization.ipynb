# **1. What is Data Visualization?**
Data Visualization is a graphical representaion of information/data. It helps **communicate insights, detect trends and simplify complex data**

## **Why is it important?**
- Makes Pattern easier to spot.
- Used in dashboards, reports, and presentations.
- Helps non-technical stakeholders understand data.

# **2. Introduction to Matplotlib**

```python
import matplotlib.pyplot as plt

students = ['Ravi', 'Sneha', 'Arjun', 'Diya', 'Rishi']
scores = [88, 79, 62, 91, 55]
```

## **1. Line Plot (For Trends)**

```python
plt.plot(students, scores) # plot(x, y)
plt.title("Student Scores")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

## **2. Bar Chart**
Great for comparing categories

```python
plt.bar(students, scores, color= 'Skyblue')
plt.title("Student Scores")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

## **3. Pie Chart**
Shows Proportion

```python
labels = ['Passed', 'Failed']
values = [3, 2]
plt.pie(values, labels = labels, autopct='%1.1f%%')
plt.title("Pass/Fail distribution")
plt.show()
```

# **2. Introduction to Seaborn**

### Load and preview: 

```python
import seaborn as sns
import pandas as pd

df=pd.read_csv("students.csv")
```

### **1. Histogram**
Shows distribution of scores. Helps detect skewness, central tendency, and outliers.

```python
sns.histplot(df['Score'], bins = 5, kde = True)
# Bins = x axis is divided into bins which help summarize the dist. of data by grouping continuous values into discrete intervals
# KDE = kernal density estimation, reveals whether the data is normally distributed skewed, multimodal or has other patterns
plt.title("Score Distribution")
plt.show()
```

### **2. Countplot(Categorical bar chart)**

```python
sns.countplot(x = 'Course', data = df)
plt.title("Number of Students per Course")
plt.xticks(rotation = 15)
plt.show
```

### **3. Boxplot(Outlier detection + Comparison)**

```python
sns.boxplot(x= 'Course', y='Score', data = df)
plt.title("Score distribution by Course")
plt.show()
```

### **4. Scatter Plot (relationship between variables)**

```python
sns.scatterplot(x='Age', y='Score', hue='Passed', data = df)
plt.title("Age vs Score with Pass Status")
plt.show()
```

```python
df = pd.read_csv("course_engagement.csv")
```

```python
sns.scatterplot(x= 'TimeSpent', y='Score', hue='Completed', style='Gender', data = df)
plt.title("Time Spent vs Score")
plt.show()
```

### **5. Correlation Heatmaps**
when to use? 
- Numerical data only
- Help detect highly correlated or redundant features

```python
corr = df.corr(numeric_only=True) # Numeric_only = True means only numeric columns are considered
sns.heatmap(corr, annot=True, cmap= 'coolwarm') # annot-> adds the actual correlation values inside each cell of the heatmap
# coolwarm-> warm= colours that show high positive correlation
        #    cool= colours that show high negative correlation, white = low/zero correlation   
plt.title("Correlation Matrix") 
plt.show()
```

### **6. Pair Plots**
A grid of scatterplot for every pair of numerical features, with histograms on the diagonal
when to use? 
- See how multiple features relate at once
- Quickly detect which variables are good predictors

```python
sns.pairplot(df, hue='Completed')
plt.show()
```
